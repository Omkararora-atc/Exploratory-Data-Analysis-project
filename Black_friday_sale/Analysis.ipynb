### Analysis of gender distribution

```python
gender_count=df["Gender"].value_counts()
print(f"Number of male users: {gender_count['M']}")
print(f"Number of female users: {gender_count['F']}")
```

```python
plt.figure(figsize = (6,6))
gender_count.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B3FF'])
plt.title('Gender Distribution')
plt.plot()
```

### Analysis of User Purchases based on Age Group

```python
age_count = df["Age"].value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.barplot(x=age_count.index, y=age_count.values, palette='Set1')
legend=False
plt.title('User Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('User Count')
plt.show()
```

### Observations
* The age group of 26-35 years has the highest number of users, followed by the age group of 36-45 years.
* The age group of 36-45 years has a moderate number of users, while the age group of 55+ and 0-17 years has the lowest number of users.

```python
data=df.groupby(["Gender",'Age']).size().reset_index(name='count')
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=data, x='Age', y='count', hue='Gender', palette='Set1')
plt.title('User Count by Gender and Age Group')
plt.xlabel('Age Group')
plt.ylabel('User Count')
plt.legend(title='Gender')
plt.show()
```

### Observations
* the percentage of female users is apx 25%
* the percentage of male users is apx 75%.
* The age group of 26-35 years has the highest number of users, followed by the age group of 36-45 years.
* The age group of 55+ and 0-17 years has the lowest number of users
* Male user of age group 26-35 has the highest count i.e above 160000

### Analysis of User Purchases based on City Category

```python
data=df["City_Category"].value_counts()
print(f"Number of users from City A: {data['A']}")
print(f"Number of users from City B: {data['B']}")
print(f"Number of users from City C: {data['C']}")
```

```python
data.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B3FF', '#99FF99'])
plt.title('City Category Distribution')
plt.show()
```

```python
data_city=df.groupby(['City_Category','Gender','Age']).size().reset_index(name='count')
```

```python
data_city.head()
```

```python
pivot_table = data_city.pivot_table(index='City_Category', columns=['Gender', 'Age'], values='count', fill_value=0)

plt.figure(figsize=(16,6))
sns.heatmap(pivot_table,cmap='YlGnBu')
plt.title('User Count Heatmap by City, Gender, and Age')
plt.xlabel('Gender, Age')
plt.ylabel('City Category')
plt.show()
```

```python
grouped = df.groupby(['Age', 'City_Category', 'Gender']).size().reset_index(name='count')
plt.figure(figsize=(14,7))
sns.barplot(data=grouped, x='Age', y='count', hue='City_Category', palette='Set2',
            hue_order=['A', 'B', 'C'])
plt.title('User Count by Age Group, City Category, and Gender')
plt.xlabel('Age Group')
plt.ylabel('User Count')
plt.legend(title='City Category')
plt.show()
```

### Observations
* The majority of users are from City B, followed by City C and City A.
* The age group of 26-35 years has the highest number of users in all three cities.
