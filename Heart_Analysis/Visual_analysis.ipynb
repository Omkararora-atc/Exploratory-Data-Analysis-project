### Visualizing the distribution of target variable

```python
sns.displot(df['Age'], kde = True, color = 'blue')
plt.show()
```

```python
sns.displot(df["RestingBP"] , color = 'green', kde = True)
plt.show()
```

```python
sns.displot(df["Cholesterol"], kde = True, color = 'red')
```

```python
sns.displot(df["MaxHR"], kde = True, color = 'blue')
```

### **Observations:**
* In the age column the maximum count is in the range of 50-60 years.
* The maximum person have the Cholesterol level in the range of 0.
* The maximum person have the RestingBP in the range of 120-140.
* The minimum person have the MaxHR in the range of 60-80.

### Analyzing the distribution of categorical variables

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f', figsize = (6,6))
```

### Observations:
* The maximum number of people have the ChestPainType as ASY.
* the minimum number of people have the ChestPainType as TA.
* the distribution of Gender men have more count
* the maximum number of people have the ST_Slope as Flat.
